```python
from hiive.mdptoolbox import mdp, example
import hiive
import matplotlib.pyplot as plt
from gym.envs.toy_text.frozen_lake import generate_random_map
import json
import numpy as np

random_maps = {}
np.random.seed(0)

# for s in sizes:
#     random_map = generate_random_map(size=s, p=0.9)
#     random_maps[str(s)]  = random_map

# json_object = json.dumps(random_maps, indent=4)

# with open("maps.json", "w") as outfile:
#     outfile.write(json_object)

with open('maps.json', 'r') as openfile:

    # Reading from json file
    random_maps = json.load(openfile)
```

**Value Iteration**

```python
# change state size here to generate different plots
state_size = 10
iterations = []
times  = []
rewards = []

for _ in range(10):
    random_map = generate_random_map(size=state_size, p=0.9)
    P, R = example.openai("FrozenLake-v1", desc=random_map)

    vi = mdp.ValueIteration(P, R, 0.9)
    stats = vi.run()
    iterations.append(vi.iter)
    times.append(vi.time)
    reward = [stat['Reward'] for stat in stats]
    rewards.append(reward)
    
print('mean iteration: ', sum(iterations)/len(iterations))
print('max iteration: ', max(iterations))
print('min iteration: ', min(iterations))

print('mean time: ', round(sum(times)/len(times), 4))
print('max time: ', round(max(times), 4))
print('min time: ', round(min(times), 4))

#         plt.plot(range(1, vi.iter+1), reward, label=f"State size {s}x{s}")
mean_list = []
max_list = []
min_list = []
for i in range(20):

    acc = 0
    count = 0
    max_r = 0
    min_r = 10
    for r in rewards:

        if i < len(r):
            acc += r[i]
            count += 1
            if r[i] > max_r:
                max_r = r[i]
            if r[i] < min_r:
                min_r = r[i]
    if count == 0: continue
    mean_list.append(acc/count)
    max_list.append(max_r)
    min_list.append(min_r)

plt.plot(range(1, len(mean_list)+1), mean_list, label=f"size {state_size}x{state_size} (average)")
plt.fill_between(range(1, len(mean_list)+1), min_list, max_list, alpha=0.2)

plt.legend()
plt.xlabel("Iteration")
plt.ylabel("Reward")
plt.title("Rewards vs Iterations in Value Iteration (Frozen Lake)")
```

**Policy Iteration**

```python
iterations = []
times  = []
sizes = range(10, 60, 10)

for s in sizes:
    print(s)
    
    P, R = example.openai("FrozenLake-v1", desc=random_maps[str(s)])

    pi = mdp.PolicyIteration(P, R, 0.9, max_iter=5)
    stats = pi.run()
    iterations.append(pi.iter)
    times.append(pi.time)
    print(stats[-1]['Reward'])
    reward = [stat['Reward'] for stat in stats]
    plt.plot(range(1, pi.iter+1), reward, label=f"State size {s}x{s}")
    # stats
plt.legend()
plt.xlabel("Iteration")
plt.ylabel("Reward")
plt.xticks(range(1, 6))
plt.title("Rewards vs Iterations in Policy Iteration (Frozen Lake)")
```

**Q Learning**

```python
iterations = []
times  = []
sizes = range(10, 60, 10)

for s in sizes:
    print(s)
    
    P, R = example.openai("FrozenLake-v1", desc=random_maps[str(s)])

    ql = mdp.QLearning(P, R, 0.9, n_iter=100000)
    stats = ql.run()
#     iterations.append(ql.iter)
    times.append(ql.time)
    print(stats[-1]['Reward'])
    reward = [stat['Reward'] for stat in stats]
    plt.plot(range(1, 10001), reward, label=f"State size {s}x{s}")
    # stats
plt.legend()
plt.xlabel("Iteration")
plt.ylabel("Reward")
# plt.xticks(range(1, 6))
plt.title("Rewards vs Iterations in Q-Learning (Frozen Lake)")
```
